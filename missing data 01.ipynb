```python
#percentage of null values in each column
null_var = df.isnull().sum()/df.shape[0] *100
null_var
```

```python
waste_columns = null_var[null_var > 17].keys() #keys are columns name.
waste_columns
```

```python
df2_drop = df.drop(columns = waste_columns)
df2_drop.head(6)
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0] * 100, 
          df3_drop[var].value_counts()/df3_drop.shape[0] * 100], axis=1,
         keys=[var+'_org', var+'clean'])
```
