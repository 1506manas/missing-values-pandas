```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df.head(6)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
miss_value = df.isnull().sum()/df.shape[0]*100
miss_value
```

```python
miss_value_clm_20 = miss_value[miss_value > 20].keys()
miss_value_clm_20
```

```python
df2_drop = df.drop(columns = miss_value_clm_20)
df2_drop.shape
```

```python
df3_num = df2_drop.select_dtypes(include = ['int64','float64'])
df3_num.head(6)
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis = 1)]
```

```python
df3_num.isnull().sum()
```

```python
miss_value_df3 = [var for var in df3_num.columns if df3_num[var].isnull().sum() > 0]
miss_value_df3
```

```python
plt.figure(figsize = (10,10))
for i, var in enumerate(miss_value_df3):
    plt.subplot(2,2,i+1)
    sns.histplot(df3_num[var], bins = 20)
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.head(10)
```

```python
#df4_num_mean.isnull().sum()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(miss_value_df3):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```

```python
df_concat = pd.concat([df3_num[miss_value_df3], df4_num_mean[miss_value_df3],df5_num_median[miss_value_df3]], axis = 1)
df_concat[df_concat.isnull().any(axis=1)]
```
